```python
def is_early_retired(row):
    country = row["country"]
    age = row["age"]
    gender = "Women" if row["female"] == 1 else "Men"
    if country in retirement_age:
        retirement_age_country = retirement_age[country]["Early age"][gender]
        if row["ep005_"] == 1:  # Retired
            if age < retirement_age_country:
                return 1  # Early retired
            else:
                return 0  # Retired but don't know
        elif row["ep005_"] == 2:
            if age < retirement_age_country:
                return -2  # Employee and age less than retirement age of country
            else:
                return -1  # Employee and age greater than the retirement age of their country
        else:
            return -3 #Other unemployment/sick/homemaker
    else:
        return -4  # Country not found in retirement_age dictionary

# Apply the function to create the new column
df["early_retired"] = df.apply(is_early_retired, axis=1)
df.head()
```
